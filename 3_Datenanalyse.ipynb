## Erstellung eines DTM 

Die Spalten eines DTM sind alle vorkommenden Wörter, die Indizes sind die Knowledbase-Artikel und die Zelleninhalte sind die Häufigkeiten des Vorkommens eines Wortes in einem Artikel

```python
from sklearn.feature_extraction.text import CountVectorizer

def dtm(corpus):    
    #print(corpus)  
    #Wenn der Zelleninhalt eine Liste ist, wird dieser in ein String umgewandelt
    for s in corpus.index:
        if isinstance(corpus[s], list):
            text = ', '.join(corpus[s])
            corpus[s] = text
                
                
                
    #print(corpus)           
    # Zählt jedes Wortvorkommen und stellt diese in einer Matrix da
    cv = CountVectorizer()    
    
    data_cv = cv.fit_transform(corpus)
    data_cv
    
    data_dtm = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
    data_dtm.index = corpus.index

    data_dtm.to_pickle("data/dtm.pkl")
    return data_dtm
          

# ******************************************************************************+
# Laden des Corpus
file = open("data/clean_Resolution.pkl", "rb")
kb = pickle.load(file)    
    
data_dtm = dtm(kb)
data_dtm.iloc[1:10,50:70]
```

```python
file = open("data/dtm.pkl", "rb")
dtm_kb = pickle.load(file)    


def del_useless_words(dtm):
    counted_words = dtm.sum(axis=0)
    wordlist = counted_words.index.tolist()
    
    index = 0
    dropwords = []
    for el in counted_words:
        if el < 3:
            dropwords.append(wordlist[index])
        index +=1
    
    dtm = dtm.drop(columns=dropwords, axis=1)
    data_dtm.to_pickle("data/clean_dtm.pkl")
    return dtm

del_useless_words(dtm_kb)
```

## Top X Wörter

Für jeden Knowledgebase Artikel werden die Wörter gezählt und die Top X ausgegeben

```python
#Ermittelt die Top x Wörter für jeden Artikel und Speichert diese in ein Tupel (wort, anzahl)
# Gibt ein Dictonary zurück und nimmt ein dtm entgenen
def topX(column, dateiname = "topX", x = 100):
    
    top_dict = {}
    top_list = []
    for c in column:
        top = column[c].sort_values(ascending=False).head(x)
        top_list.append(top)
        top_dict[c]= list(zip(top.index, top.values))    
        top_df = pd.DataFrame(top_dict)
        
    indizes = kb.index.tolist()#!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    toplist = []
    
    for element in top_list:
        toplist.append(element.index.tolist())
        
        
    temp = []
    indizes = kb.index.tolist()#!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    index = 0
    for i in toplist:
        temp.append(i)
        index +=1
    
    toplist = pd.Series(temp, index = indizes)
    toplist.to_pickle("data/top100Series.pkl")
    return top_df, toplist


# ***********************************************
file = open("data/clean_dtm.pkl", "rb")
cl_dtm = pickle.load(file)    


topx_dataframe, liste = topX(cl_dtm.transpose())

#topx_dataframe
#liste
cl_dtm.transpose()
```

## Unbedeusten Top X Wörter

Für jeden Artikel werden die unbedeutesten Wörter herausgesucht. Die könnten aus der Liste gelöscht werden.

```python
#Ermittelt die Top x Wörter für jeden Artikel und Speichert diese in ein Tupel (wort, anzahl)
# Gibt ein Dictonary zurück und nimmt ein dtm entgenen
def lessX(column, dateiname = "lessX", x = 50):
    
    top_dict = {}

    for c in column:
        top = column[c].sort_values(ascending=True).head(x)
        # top.index = Wortname
        top_dict[c]= list(top.index)    
    
    return top_dict
    
col = data_dtm.transpose()
top_dict = lessX(col)
top_dataFrame = pd.DataFrame(top_dict)
top_dataFrame.transpose()
```

```python
from collections import Counter

def mostCommonWord(dtm):
    
    # Ermittlung der Top Wörter für alle Artikel    
    top_dict = topX(dtm)[0]
    
    words = []
    for i in dtm:
        top = [word for (word, count) in top_dict[i]]
        
        for t in top:
            words.append(t)
        
    return words

def CommonStopWords(data, percentage = 0.9):
    # Liste der eigens definierten Stop-Wörter
    add_stop_words = [word for word, count in Counter(mostCommonWord(data)).most_common() if count >= len(data)*percentage]

    return add_stop_words



col = data_dtm.transpose()
print(CommonStopWords(col))
```
