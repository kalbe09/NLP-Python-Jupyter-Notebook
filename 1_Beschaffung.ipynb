# Beschaffung eines Datensatzes

Lesen einer CSV-Datei "kb" mit den Spalten: Topic, ID, Question, Reference, Enviroment, Resolution
Stellt eine Knowledgbase mit Incidents da, um auf diese das Natural Language Processing anzuwenden.

Es werden alle Datensätze entfernt, die ein NaN enthalten. 
Die Indizes müssen mit den IDs in der Spalte ID getauscht werden, ansonsten können die Url's der Knowledbase-Artikel nicht aufgerufen werden.

```python
import re
import pandas as pd
import pickle
```

```python
def read_csv(filename):
    kb = pd.read_csv('' + filename + '.csv', sep=';')

    # this will drop all rows with at least one NaN
    kb = kb.dropna(how='any')
    #print(kb)
    # change indizes with the values in column ID
    id = kb['ID'].tolist()
    
    # filter out the id by substracting 'Anser ID'
    for i in range(len(id)):
        id[i] = int(re.sub('Answer ID ', '', id[i]))
    
    # save the work in a new dataframe, drop column ID and pickle it for later use
    kb = pd.DataFrame(kb.values.tolist(), index=id,
                      columns=['Topic', 'ID', 'Question', 'Reference', 'Enviroment', 'Resolution'])
    kb = kb.drop(columns='ID', axis=1)
    
    # remove later, for faster working
    kb = kb[:30]
    
    kb.to_pickle("data/knowledgebase.pkl")
    kb.to_csv("r/knowledgebase.csv", sep=";")
    return kb

kb = read_csv('orig_data/kb')
```
